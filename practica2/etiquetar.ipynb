```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from IPython.display import clear_output
```

```python
def divide_image(image_path, width, height):
    # Lee imagen original
    image = cv2.imread(image_path)
    
    h, w, _ = image.shape
    
    # Divisiones en horizontal y vertical
    num_horizontal = w // width
    num_vertical = h // height
    
    regions = []
    
    # Regiones
    for i in range(num_vertical):
        for j in range(num_horizontal):
            y_start = i * height
            y_end = (i + 1) * height
            x_start = j * width
            x_end = (j + 1) * width
            
            region = image[y_start:y_end, x_start:x_end]
            regions.append(region)
            
    return regions
```

```python
def etiquetar():
    # Imagen original
    image_path = "./imagen1.webp"
    
    # Dimensiones de las regiones
    width = int(input("Ancho del cuadro: "))
    height = int(input("Alto del cuadro: "))
    
    # Obtener las regiones de la imagen
    regions = divide_image(image_path, width, height)
    
    # Mostrar cada regi√≥n y guardarla
    for i, region in enumerate(regions):
        plt.imshow(cv2.cvtColor(region, cv2.COLOR_BGR2RGB))
        plt.show()
        
        folder_name = input("Etiqueta de imagen: ") # Puede ser C (ciudad) / A (animal) / P (persona)
        foler_name = f"./DATASET/{folder_name}"

        if not os.path.exists(folder_name):
            os.makedirs(folder_name)
        
        # Guardar la imagen
        image_name = f"{i + 1}.jpg"
        cv2.imwrite(os.path.join(folder_name, image_name), region)
        
        print(f"Imagen {image_name} guardada en la carpeta {folder_name}")
        clear_output(wait=True)
```

```python
etiquetar()
```
